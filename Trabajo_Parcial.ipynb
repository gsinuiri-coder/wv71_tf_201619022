**Importar las bibliotecas para el proyecto**

```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
```

**Selección aleatoria de las coordenas y la cantidad de almacenes y puntos de entrega**

1.   Se elige de forma aleatoria y no repetida 5100 coordenadas que seria en el caso extremo la elección de 100 almacenes y 5000 puntos de entrega. 
2.   Se elige de forma aleatoria la cantidad de almacenes y puntos de entrega.
3.   De las 5100 coordenadas se selecciona la cantidad de almacenes y desde este ultimo la cantidad de puntos de entrega.
4.   Se genera los archivos "almacenes.csv" y "puntos_entrega.csv".








```python
points = np.random.choice(range(20000), (5100, 2), replace=False)
cantidad_almacenes = npr.randint(50, 100)
cantidad_puntos_entrega = npr.randint(2500, 5000)

print("Almacenes: ", cantidad_almacenes)
print("Puntos de entrega: ", cantidad_puntos_entrega)

almacenes = points[:cantidad_almacenes]
puntos_entrega = points[cantidad_almacenes: cantidad_almacenes + cantidad_puntos_entrega]

np.savetxt('almacenes.csv', almacenes, fmt="%i", delimiter=",", header="x,y", comments="")
np.savetxt('puntos_entrega.csv', puntos_entrega, fmt="%i", delimiter=",", header="x,y", comments="")
```

**Separar coordenas (x, y) para usarlo con la biblioteca de matplotlib (parte visual)**

```python
x_almacen = almacenes[:,0]
y_almacen = almacenes[:,1]

x_puntos_entrega = puntos_entrega[:,0]
y_puntos_entrega = puntos_entrega[:,1]
```

**Resultado parte visual**

```python
fig, ax = plt.subplots()

ax.set_xlabel("x", fontsize=30)
ax.set_ylabel("y", fontsize=30)
ax.set_title('Almacenes y Puntos de entrega', fontsize=30)

ax.grid(True)
fig.set_size_inches(20, 15)

ax.plot(x_almacen, y_almacen, 'ro', label="Almacen", markersize=10)
ax.plot(x_puntos_entrega, y_puntos_entrega, 'go', label="Punto de entrega")

ax.legend()
plt.show()
```
